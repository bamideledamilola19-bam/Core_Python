```python
#QUESTION 1. SWAP FIRST AND LAST ELEMENT OF A LIST. MODIFY A GIVEN LIST SUCH THAT THE FIRST AND LAST ELEMENT ARE SWAPPED.

fruits = ['apple', 'cherry', 'banana', 'lemon']
print(fruits)
#swap first and last element
fruits[0], fruits[-1] = fruits[-1], fruits[0]
print(fruits)
```
